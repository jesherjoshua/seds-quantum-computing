# Some Common Gates in Qiskit


```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
import matplotlib.pyplot as plt
```

```python
qr = QuantumRegister(1)
cr = ClassicalRegister(1)
```

```python
circ = QuantumCircuit(qr, cr)
```

```python
circ.measure(0, 0)
```

```python
circ.draw("mpl", scale=3)
```

```python
simulator = BasicAer.get_backend("qasm_simulator")
```

```python
job = execute(circ, simulator, shots=1000)
```

```python
counts = job.result().get_counts()
```

```python
plot_histogram(counts)
```

## A neat function to view the results ðŸ˜Œ

```python
def view_results(circuit, simulator):
    job = execute(circ, simulator, shots=1000)
    counts = job.result().get_counts()
    print(f"Counts for 1000 shots: {counts}")
    return plot_histogram(counts)
```

## Pauli Gate

**Pauli's X-gate is also often called a NOT-gate**

`circ.x(index_of_qubit)`


```python
circ.x(0)
circ.measure(0, 0)
view_results(circ, simulator)
```

```python
circ.draw('mpl')
```

## Hadamard Gate

`circ.h(index_of_qubit)`


```python
circ.h(0)
circ.measure(0, 0)
view_results(circ, simulator)
```

```python
circ.draw('mpl')
```

## CNOT Gate

The CNOT gate is two-qubit operation, where the first qubit is usually referred to as the control qubit and the second qubit as the target qubit. Expressed in basis states, the CNOT gate:

- leaves the control qubit unchanged and performs a Pauli-X gate on the target qubit when the control qubit is in state âˆ£1âŸ©
- leaves the target qubit unchanged when the control qubit is in state âˆ£0âŸ©.

`circ.cx(index_of_control_qubit,index_of_target_qubit)`


```python
circ = QuantumCircuit(2, 2)
# circ.x(0)
circ.cx(0, 1)
circ.measure(0, 0)
circ.measure(1, 1)
circ.draw("mpl")
view_results(circ, simulator)
```

```python
circ.draw("mpl")
```

# Logic Gates using Qiskit


```python
def test(circuit):
    backend = BasicAer.get_backend("qasm_simulator")
    job = execute(circuit, backend, shots=1)
    result = job.result()
    return list(result.get_counts().keys())[0]
```

## Classical AND Gate

```python

print("A\tB\tY")
l = [(0, 0), (0, 1), (1, 0), (1, 1)]
for a, b in l:
    circ = QuantumCircuit(3, 1)
    if a != 0:
        circ.x(0)
    if b != 0:
        circ.x(1)
    circ.ccx(0, 1, 2) # Toffoli Gate
    circ.measure(2, 0)
    print(f"{a}\t{b}\t{test(circ)}")

circ.draw('mpl')
```

## Classical OR Gate

```python
print("A\tB\tY")
l = [(0, 0), (0, 1), (1, 0), (1, 1)]
for a, b in l:
    circ = QuantumCircuit(3, 1)
    if a != 0:
        circ.x(0)
    if b != 0:
        circ.x(1)
        
    circ.ccx(0,1,2)
    circ.cx(0,2)
    circ.cx(1,2)

    circ.measure(2,0)
    print(f"{a}\t{b}\t{test(circ)}")
circ.draw('mpl')
```

## Classical NOT Gate

```python
print("A\tY")
l = [0,1]
circ = QuantumCircuit(1, 1)
for a in l:
    circ.x(0)
    

    circ.measure(0,0)
    print(f"{a}\t{test(circ)}")
circ.draw('mpl')
```

## XOR Gate

```python

print("A\tB\tY")
l = [(0, 0), (0, 1), (1, 0), (1, 1)]
for a, b in l:
    circ = QuantumCircuit(3, 1)
    if a != 0:
        circ.x(0)
    if b != 0:
        circ.x(1)
    circ.cx(0, 1)
    circ.measure(1, 0)
    print(f"{a}\t{b}\t{test(circ)}")

circ.draw('mpl')
```
